## Nation-wise Contribution

*Let's see which country has the most participation to the sport.*

```python
value_counts = df['nationality'].value_counts()
nat = pd.DataFrame({'Nationality':value_counts.index , '# of Players' : value_counts.values})
nat_df = nat.head(20)
nat_df
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data = nat_df, x = 'Nationality' , y = '# of Players')

for bars in ax.containers:
    ax.bar_label(bars)
```

*It can be clearly seen that England has the most number of players.*

## Top teams

The top teams will be having the maximum sum of the overall ratings of the players.

```python
final_team=df[['short_name','age','club','player_positions','overall']]
final_team.sort_values(by='player_positions')
pos_play=final_team.groupby('player_positions').apply(lambda x:np.max(x['overall'])).reset_index(name='Overall Score')
player_pos=pd.merge(final_team,pos_play,how='inner',left_on=['player_positions','overall'],right_on=['player_positions','Overall Score'])
pos_best=player_pos[['short_name','club','age','player_positions','Overall Score']]
pos_best.head(20)
```

```python
ovr_team = pos_best.groupby('club', as_index = False)['Overall Score'].sum().sort_values(by = 'Overall Score',ascending = False )
sns.set(rc={'figure.figsize':(20,10)})
ax = sns.barplot( data = ovr_team.head(10) , x = 'club', y = 'Overall Score')
for bars in ax.containers:
    ax.bar_label(bars)
ovr_team
```

*The top teams are Real Madrid followed by FC Bayern Munchen and Napoli*


## Clubs having the greatest potential.

```python
final_team=df[['short_name','age','potential','player_positions','club']]
final_team.sort_values(by='age')
pos_play=final_team.groupby('player_positions').apply(lambda x:np.max(x['potential'])).reset_index(name='Potential')
player_pos=pd.merge(final_team,pos_play,how='inner',left_on=['player_positions','potential'],right_on=['player_positions','Potential'])
pot_play=player_pos[['short_name','club','age','player_positions','Potential']]
pot_play.head(20)
```

```python
pot_team = pot_play.groupby('club', as_index = False)['Potential'].sum().sort_values(by = 'Potential',ascending = False )
sns.set(rc={'figure.figsize':(22,5)})
ax = sns.barplot( data = pot_team.head(10) , x = 'club', y = 'Potential')
for bars in ax.containers:
    ax.bar_label(bars)
pot_team
```

*Thus it can be seen Real Madrid has the greatest potential.*
